# OLS Derivations

## 1. Assumptions
- **Linear model** : $ y = X\beta + \epsilon $
- **Exogeneity:**   $\ \ \mathbb{E}_{\mathbb{X}}[\epsilon \ | \ X] = 0 \ \ $ (model makes no systematic mistakes)
- **homoscedasticity**: $ Var(\epsilon | X) = \sigma^2 I_n $
- **random sampling**: $(x_i, y_i)$ are independent
- $X$ **is fixed (non-random)**
- $X$ **has full rank**: $x_i \ $ (features) are uncorrelated


## 2.OLS Normal Equation 

### 2.1 Formulating the Optimization Problem

#### a. Model Formulation:
Linear regression assumes the response is a linear combination of the predictors:
$$ y = X\beta + \epsilon$$
Where:
- $ X \in \mathbb{R}^{n \times (p+1)}$ : design matrix (first column is $ 1_{n}$)
- $ y \in \mathbb{R}^{n}$ : target vector
- $ \beta \in \mathbb{R}^{p+1}$ : coefficient vector
- $ \epsilon \in \mathbb{R}^{n}$: eror term
- $n$ : number of observations
- $p$ :  number of features

#### b. Prediction
Since noise $\epsilon$ (irreducible error) can't be observed. we can try to predict y using only X by finding $\hat{\beta}$:
$$
    \hat{y} = X\hat{\beta}
$$



#### c. Loss Function:
To measure how much off our prediction is we need a measure that takes both negative and positive offsets  
F that we can use the quadratic loss function $\ell_2 $ :  
$$ 
\ell_2(\hat{y}, y) = \|\hat{y} - y\|_2^2
$$
We will also define residual $e$ as:  
$$
    e = y - \hat{y}
$$
Therefore our goal is to find $ \hat{\beta}$ an estimator of $\beta$ that fits the best meaning the one that minimizes $\|e\|_2^2 \\[3em]$

Since we have $n$ observations the goal becomes equivalent to minimizing the Residual Sum of Squares RSS: 
$$ \ \ RSS(\beta) = \sum_{i=0}^{n}e_i^2 = e^Te = (y - X\beta)^T(y - X\beta)$$  

#### d. Optimization Problem Formulation
Then we can formulate the problem as:  
$$ \hat{\beta} = \arg\min_{\beta} (y - X\beta)^T(y - X\beta) \\[3em]$$  
Which is the **Ordinaryy Least Squares(OLS)** problem

### 2.2 Solving the OLS Optimization Problem 

#### a. Showing That $X^TX$ is Positve Definite:
we have: $\forall x \in \mathbb{R}^{p}$
$$ 
\begin{aligned}
 x^T(X^TX)x &= (Xx)^T(Xx) \\ 
 &= \|Xx\|_2 >= 0
\end{aligned}
$$
thus X^X is postive semi definite 
but since we assumed $X$ has full rank then we have: $$ det(X) \neq 0 \Longrightarrow det(X^TX) = det(X^T)det(X) = det(X)^2 \neq 0 $$ 
thus $X^TX$ is invertible then $X^TX$ is **positive definite**

#### b. Finding Critical Points
$$
\begin{aligned}
\frac{\partial RSS}{\partial \beta}(\beta_c) &= 0 \\
\frac{\partial}{\partial \beta} (y^Ty - 2y^TX\beta_c + \beta_c^TX^TX\beta_c) &= 0 \\
-2y^TX + 2X^TX\beta_c &= 0 \\
(X^TX)\beta_c &= X^Ty \\
\beta_c &= (X^TX)^{-1}X^Ty \\
\end{aligned}
$$

#### c. Determining the Nature of the Critical Point
We determine the nature of $ \beta_c$ by checking the second partial derivative 
$$ \frac{\partial^2 RSS}{\partial \beta \, \partial \beta^T}(\beta_c) = 2X^TX  $$
Which is positive definite thus $\beta_c$ is a **Unique Global Infinimum** and 
$$\hat{\beta_{\mathrm{ols}}} = \beta_{\mathrm{c}} =(X^TX)^{-1}X^Ty $$
And thus 
$$ \hat{y} = X(X^TX)^{-1}X^Ty $$ 
As the best prediction of $y$ based on $X$

### 2.3 Geometric Intuition

#### a. Link with Orthogonality

by setting $$ H = X(X^TX)^{-1}X^T $$
we obtain $$\hat{y} = Hy $$
where  **$H$** oftend called **the Hat Matrix** is the **Orthogonal Projection Matrix** onto the vectorial subspace spanned by $X_i$ the vector columns of $X$ which corresponds to the result of **Orthogonal Projection Theorem**

#### b. Orthogonal Projection Theorem
Let $(V, .)$ an inner product space (a vectorial space with an inner product). then for any vector $y$ there exists a unique vector $\hat{y} \in V$ such that:  
$$\hat{y} = \arg\min_{v \in V}\|y - v\| $$   
and this vector is precisely the orthogonal projection of $y$ onto $V$:  
$$ \hat{y} = \operatorname{proj}_{V}(y) $$

#### c. Alternative Derivation of $\hat{\beta_{\mathrm{ols}}}$ Using Orthogonality
The same result can be obtained directly by applying the **Orthogonal Projection Theorem** as follows:
$$ \begin{aligned}
    & & X . e &= 0 \\
    &\Longrightarrow& \quad X^T(y - X\hat{\beta_{\mathrm{ols}}}) &= 0 \\
    &\Longrightarrow& \quad X^TX\hat{\beta_{\mathrm{ols}}} &= X^Ty \\
    &\Longrightarrow& \quad \hat{\beta_{\mathrm{ols}}} &= (X^TX)^{-1}X^Ty
     \end{aligned}
$$
$

#### d.More About the Hat Matrix $H$:
H has some nice properties:
- **symmetric**:
  $$
  \begin{aligned}
  H^T &= (X(X^TX)^{-1}X^T)^T \\
      &= X(X^TX)^{-T}X^T \\
      &= X(X^TX)^{-1}X^T \\
      &= H
  \end{aligned}
  $$
- **idomptent**  
  $$
  \begin{aligned}
  H^2 &= X(X^TX)^{-1}X^T X(X^TX)^{-1}X^T\\
      &= X(X^TX)^{-1}(X^TX)(X^TX)^{-1}X^T \\
      &= X(X^TX)^{-1}X^T \\
      &= H
  \end{aligned}
  $$
  these first two properties are by definition since it is an **Orthogonal Projection Matrix**
  
- $rank(H)$:
  $H$ has rank $p$
- $Trace(H)$:
    Using the circular property of trace:
  $$
  Trace(ABC) = Trace(CAB)
  $$
  we have:
  $$
  \begin{aligned}
  Trace(H) 
  &= Trace(X(X^TX)^{-1}X^T)\\
  &= Trace(X^TX(X^TX)^{-1}) \\
  &= Trace(I_p) \\
  &= p
  \end{aligned}
  $$
  -**Eigenvalues**:
   
  Let $\lambda_i$ be an eigenvalue of H, and $v_i$ be its corresponding non-zero vector then by using the idomptency property:
  $$
  \begin{aligned}
  Hv_i &= \lambda_iv_i \\
  Hv_i &= H^2v_i = H(Hv_i) = \lambda_i Hv_i = \lambda_i^2v_i
  \end{aligned}
  $$
  thus
  $$
  \lambda_i^2v_i = \lambda_iv_i
  $$
  since $v_i$ is a non-zero vector we get 
  $$
  \lambda_i^2 = \lambda_i \Longrightarrow \quad \lambda_i = 1 \; or \; \lambda_i = 0
  $$
  -**Information Loosing**:  
  Othogonal Projection Matrices are generally not invertible because they doesn't preserve enough information
  (e.g two distinct points can have the same orthogonal projection on a plane)

### 3.Properties of $\hat{\beta_{\mathrm{ols}}}$

- **$\hat{\beta_{\mathrm{ols}}}$ is unbaised**:  
  using the **Exogeneity** assumption we have: :
  $$
  \begin{aligned}
  \mathbb{E}[\epsilon] &= \mathbb{E}[\mathbb{E}_{\mathbb{X}}[\epsilon | X]] \\
  &= \mathbb{E}[0] \\
  &= 0 \\[1em]
\mathbb{E}[\hat{\beta_{\mathrm{ols}}}] &= \mathbb{E}[(X^TX)^{-1}X^Ty] \\
    &= (X^TX)^{-1}X^T \mathbb{E}[X\beta + \epsilon] \\
  &= (X^TX)^{-1}X^TX\mathbb{E}[\beta] + (X^TX)^{-1}X^T\mathbb{E}[\epsilon] \\ 
  &= \mathbb{E}[\beta] + (X^TX)^{-1}X^T . 0 \\
  &= \mathbb{E}[\beta]
    \end{aligned}
  $$

- **variance of $\hat{\beta_{\mathrm{ols}}}$**:  
  using the **homoscedasticity** assumption we have:

  
  $$ \begin{aligned}
  Var(\epsilon) &= \mathbb{E}[Var_{X}(\epsilon | X)] + Var(\mathbb{E}_X[\epsilon | X]) \\ 
  &= \mathbb{E}[\sigma^2I_n] + Var(0) \\
  &= \sigma^2I_n \\[2em]
  Var(y) &= Var(X\beta + \epsilon)\\ &= Var(\epsilon) \\ &= \sigma^2 I_n \\[2em]
  Var(\hat{\beta_{\mathrm{ols}}}) &= \operatorname{Var}((X^TX)^{-1}X^Ty) \\
  &=(X^TX)^{-1}X^T \,\operatorname{Var}(y) \, ((X^TX)^{-1}X^T)^T \\
   &= (X^TX)^{-1}(X^TX)(X^TX)^{-1} \sigma^2 I_n  \\
   &=(X^TX)^{-1} \sigma^2
  \end{aligned} $$

- $\beta_{\mathrm{ols}}$ is **consistent** $ \\[1em] $:
  
which is the same as saying $\hat{\beta_{ols}}$ converges in probabilty:
$$
\begin{aligned}
\hat{\beta_{\mathrm{ols}}} &\stackrel{\mathbb{P}}{\longrightarrow} \beta \\
\iff\;&
\forall \epsilon>0,\;
\lim_{n\to\infty}
\mathbb{P}\!\left(
|\hat{\beta}_{ols}-\beta|<\epsilon
\right)=1
\end{aligned}
$$
but instead of using the definiton we will prove it as follows:
$$
\begin{aligned}
\hat{\beta_{\mathrm{ols}}} - \beta &= (X^TX)^{-1}X^Ty - \beta\\
&=(X^TX)^{-1}X^T(X\beta + \epsilon) - \beta \\
&= \beta - \beta + (X^TX)^{-1}X^T\epsilon \\
&= (X^TX)^{-1}X^T\epsilon\\
&= \left(\frac{X^TX}{n}\right)^{-1}
\left(\frac{X^T\epsilon}{n}\right)\\
\end{aligned} $$

we have:

$$
\begin{aligned}
\frac{X^TX}{n} 
= \frac{1}{n} \sum_{i=1}^{n}x_i x_i^T &\stackrel{\mathbb{P}}{\longrightarrow} Q = \mathbb{E}[X_iX_i^T]\;\; \text{(by weak law of large numbers)}\\
\Longrightarrow \left(\frac{X^TX}{n}\right)^{-1} &\stackrel{\mathbb{P}}{\longrightarrow} Q^{-1}
\end{aligned}
$$

and we also have:

$$
\begin{aligned}
\frac{X\epsilon}{n} 
= \frac{1}{n} \sum_{i=1}^{n}x_i \epsilon_i &\stackrel{\mathbb{P}}{\longrightarrow} \mathbb{E}[X_i\epsilon_i]\;\; \text{(by weak law of large numbers)}\\
\end{aligned}
$$

and we have: 
    
$$
 \begin{aligned}
  \mathbb{E}[X_i\epsilon] &= \mathbb{E}[\mathbb{E}_{\mathbb{X_i}}[X_i\epsilon | X]] \\
  &= \mathbb{E}[X_i\mathbb{E}_{\mathbb{X_i}}[\epsilon | X]] \\
  &= \mathbb{E}[X_i . 0] \\
  &= \mathbb{E}[0]\\
  &= 0 \\[1em]
  \end{aligned}
$$

thus we have:
    
$$
\frac{X\epsilon}{n}  \stackrel{\mathbb{P}}{\longrightarrow} 0
$$

using **Slutsky's theorem**:
$$
\begin{aligned}
 \left(\frac{X^TX}{n}\right)^{-1}
\left(\frac{X^T\epsilon}{n}\right) &\stackrel{\mathbb{P}}{\longrightarrow} Q^{-1} . 0\\
\Longrightarrow \left(\frac{X^TX}{n}\right)^{-1}
\left(\frac{X^T\epsilon}{n}\right) &\stackrel{\mathbb{P}}{\longrightarrow} 0
\end{aligned}
$$

- $\beta_{\mathrm{ols}}$ is **BLUE** **(Gauss-Markov Theorem)**

Under the previous assumptions **Gauss-Markov theorem** states that the Ordinary Least Squares (OLS) estimator is the **Best Linear Unbiased Estimator (BLUE)**.



the OLS estimator has the **smallest covariance matrix**. In other words, for any other linear unbiased estimator \(\tilde{\beta}\),

$$
\operatorname{Cov}(\tilde{\beta})
-
\operatorname{Cov}(\hat{\beta}_{\mathrm{OLS}})
\succeq 0,
$$

where $\succeq 0$ denotes that the difference is a positive semidefinite matrix.

Therefore, no other linear unbiased estimator can achieve a smaller variance than OLS. This result is known as the **Gauss-Markov theorem**, as presented in "*The Elements of Statistical Learning*" book.

- $\beta_{\mathrm{ols}}$ is **assymptotically normal**:
  $$
\sqrt{n}\left(\hat{\beta}_{\mathrm{OLS}}-\beta\right)
\;\xrightarrow{d}\;
\mathcal{N}\!\left(0,\;\sigma^2Q^{-1}\right),
$$

where

$$
Q=\lim_{n\to\infty}\frac{X^\top X}{n},
$$

assuming \(Q\) is positive definite.$ \\[2em]$

Thus the distribution of $\beta \ $(and by consequence the data $y$) can be **approximated** to be **normally distributed** for **large** numbers of $n$

## 3. OLS under the Gaussian Linear Model
In this section error(noise) will be assumed follows a multivariate-normal distribution:
$$
\epsilon \sim \mathcal{N}(0, \sigma^2I_n)
$$


### 3.1 $\beta_{\mathrm{ols}}$ as a MLE of $\beta$

#### a. Likelihood $L(\beta, \sigma^2; y, X)$
since:
$$
\epsilon \sim \mathcal{N}(0, \sigma^2I_n)
$$
then y also follows a multivariate-normal distribution:
$$
 y = X\beta + \epsilon \sim \mathcal{N}(X\beta, \sigma^2I_n)
$$
therefore 
$$
L(\beta, \sigma^2; y, X) = \frac{1}{(2\pi\sigma^2)^{\frac{n}{2}}}e^{-\frac{1}{2\sigma^2}(y - X\beta)^T(y - X\beta)}
$$
The log of the likelihood is also taken to facilitate calculations and to be used for Fischer's information:
$$
    lnL(\beta, \sigma^2; y, X) = \frac{n}{2}ln(2\pi \sigma^2) - \frac{1}{2\sigma^2}(y - X\beta)^T(y-X\beta)
$$

#### b. Maximizing Likelihood with Respect to $\beta$
-**Finding Critical Points**:  
By setting the partial derivative of the log-likelihood to 0:
$$
\begin{aligned}
\frac{\partial lnL}{\partial \beta}(\beta, \sigma^2) &= 0 \\
\frac{\partial}{\partial \beta}\left(\frac{n}{2}ln(2\pi \sigma^2) - \frac{1}{2\sigma^2}(y - X\beta)^T(y-X\beta) \right) &= 0\\
\frac{1}{\sigma^2}X^T(y - X\beta) &= 0 \\
(X^TX)\beta &= X^Ty \\
\beta &= (X^TX)^{-1}X^Ty \\
\end{aligned}
$$

-**Determining Nature of Critical Points**:   
By determinining the nature of the second partial derivative:
$$
\begin{aligned}
\frac{\partial^2 lnL}{\partial \beta \partial \beta^T}(\beta, \sigma^2) &= \frac{\partial^2}{\partial \beta \beta^T} \left(\frac{n}{2}ln(2\pi \sigma^2) - \frac{1}{2\sigma^2}(y - X\beta)^T(y-X\beta)\right) \\
&= \frac{\partial}{\partial \beta}\left( - \frac{1}{\sigma^2}X^T(y - X\beta_{\mathrm{ols}}) \right) \\
&= -\frac{1}{\sigma^2}X^TX
\end{aligned}
$$
Which is a **Negative definite** Matrix  therefore the found $\beta$ is a **Global Supremum** thus:
$$
    \hat{\beta_{\mathrm{MLE}}} = (X^TX)^{-1}X^Ty
$$
Which is the same result obtained in OLS without normality assumption


### 3.2 Efficiency of $\beta_{\mathrm{ols}}$

- $\beta_{\mathrm{ols}}$ is **efficient** $ \\[1em] $:  
according to **Cramér-Rao Lower Bound** we have:
$$
\\[1em]
(I_n(\beta_{\mathrm{ols}}))^{-1} \preceq Var(\beta_{\mathrm{ols}})
$$
where :
$$
I_n(\beta_{\mathrm{ols}}) = - \mathbb{E}\left[\frac{\partial^2 lnL(y,X, \beta_{\mathrm{ols}})}{\partial \beta \partial \beta^T}\right]
$$
Is **Fischer's Information** $ \\[2em]$

Thus by definition  $\beta_{\mathrm{ols}}$ is said to **efficient** iff its variance reaches the **Cramér-Rao** lower bound:
$$
\\[1em]
(I_n(\beta_{\mathrm{ols}}))^{-1} = Var(\beta_{\mathrm{ols}}) \\[1em]
$$


$$ \\[2em]$$
And From previous calculations:
$$
\begin{aligned}
I_n(\beta_{\mathrm{ols}}) &= - \mathbb{E}\left[\frac{\partial^2 lnL(y,X, \beta_{\mathrm{ols}})}{\partial \beta \partial \beta^T}\right] \\[0.6em]
&=  -\mathbb{E}\left[ - \frac{1}{\sigma^2}X^TX\right] \\[0.6em]
&= \frac{1}{\sigma^2}X^TX \\[0.5em]
&= (\sigma^2(X^TX)^{-1})^{-1} \\[0.5em]
&= Var(\beta_{\mathrm{ols}})^{-1}
\end{aligned}
$$
Thus the Normal Linear model estimator  $\beta_{\mathrm{ols}}$ is **efficient** meaning it's the best unbiased estimator (has the lowest variance among all unbiased estimators) which is stronger than **BLUE** (only linear models) but requires the normality assumption

## 4. Goodness of Fit
In this section we will talk about "metrics" to measure how much good of a fit is our model



A way to quantify how much good is our model is by decomposing the total variability into an explained component and an unexplained (residual) component.

### 4.1 Total Sum of Squares (TSS)

The **Total Sum of Squares** measures the total variability of the response around its sample mean:

$$
\mathrm{TSS}
=
\sum_{i=1}^{n}(y_i-\bar{y})^2
=
\|y-\bar{y}\mathbf{1}\|_2^2,
$$

where

### 4.2 Explained Sum of Squares (ESS)

The **Explained Sum of Squares** measures the variability explained by the fitted model:

$$
\mathrm{ESS}
=
\sum_{i=1}^{n}(\hat{y}_i-\bar{y})^2
=
\|\hat{y}-\bar{y}\mathbf{1}\|_2^2.
$$

### 4.3 Residual Sum of Squares (RSS)

The **Residual Sum of Squares** measures the unexplained variability:

$$
\mathrm{RSS}
=
\sum_{i=1}^{n}(y_i-\hat{y}_i)^2
=
\|y-\hat{y}\|_2^2.
$$

### 4.4 Variance Decomposition

By using the following identity

$$
y-\bar{y}\mathbf{1}
=
(\hat{y}-\bar{y}\mathbf{1})
+
(y-\hat{y}).
$$

Since the residuals are orthogonal to the fitted values,

$$
(\hat{y}-\bar{y}\mathbf{1})^T(y-\hat{y})=0,
$$

the Pythagorean theorem gives

$$
\mathrm{TSS} =\mathrm{ESS} +\mathrm{RSS}.
$$

This identity states that the total variability of the response is exactly partitioned into:

- the variability explained by the model
- the variability left unexplained by the residuals.

### 4.5 Coefficient of Determination

The coefficient of determination is defined as

$$
R^2
=
1-\frac{\mathrm{RSS}}{\mathrm{TSS}}
=
\frac{\mathrm{ESS}}{\mathrm{TSS}}.
$$

It represents the proportion of the total variance in the response that is explained by the regression model.

Some useful interpretations are:

- $R^2 = 1$: the model fits the data perfectly.
- $R^2 = 0$: the model performs no better than predicting the sample mean.
- $0 \lt R^2 \lt 1 $: the model explains a corresponding fraction of the response variability.
- Without an intercept term, $R^2$ may become negative, indicating that predicting the mean would outperform the fitted model.

But middle values like 0.6 0.8 don't say much about the model performance because they are influenced by other factors

### 5.1 Generate Synthetic data:
we will generate data as follows:
$$
\begin{aligned}
x_i &\sim \mathcal{U}([low, high]) \\ 
\epsilon_i &\sim \mathcal{N}(0, \sigma^2)\\
\end{aligned}
$$
and:
$$
y_i = ax_i + b + \epsilon_i
$$
where "high", "low", $a$, $b$, $\sigma^2$ are variable paramters 

```python
import numpy as np
```

```python
def generate_synthetic_linear_data(beta, low, high, size=None, noise=False, scale=1):
    
    if type(beta) != tuple:
        raise ValueError("beta paramter must be a tuple of length >= 2")
        
    beta = np.array(beta)
    
    rng = np.random.default_rng()
    
    X = rng.uniform(low=-3, high=3, size=(size, len(beta) - 1))
    X = np.insert(X, 0, 1, axis=1)
    
    y = X @ beta

    X = X[: , 1:]
    if noise:
        epsilons = rng.normal(loc=0, scale=scale, size=size)
        y += epsilons
        
    return (X, y)
```

### 5.2 Defining Base OLS model

```python
class BaseOLS:

    def __init__(self, random_state=None):
        
        self.random_state = random_state

    
    def fit(self, X, y):
        
        self.validate_fit_(X, y)
            
        X = self.insert_intercept_(X)
        
        self.beta_ = self.solve_(X, y)
        self.intercept_ = self.beta_[0]
        self.coef_ = self.beta_[1:]
        
        return self

    
    def predict(self, X):
        
        X = self.insert_intercept_(X)
        self.validate_predict_(X)
        return X @ self.beta_

    
    def score(self, X, y):
        
        y_hat = self.predict(X)
        residuals = y - y_hat
      
        self.rss_ = np.linalg.norm(residuals) ** 2
        self.tss_ = np.linalg.norm(y - np.mean(y)) ** 2
        
        return 1 - (self.rss_ / self.tss_)

        
    def validate_fit_(self, X, y):
        
        if (not isinstance(X, np.ndarray)) or (not isinstance(y, np.ndarray)):
            raise TypeError("Expected a NumPy array as inputs")
        
        if (not np.issubdtype(X.dtype, np.number)) or (not np.issubdtype(y.dtype, np.number)):
            raise TypeError("Numpy Array Inputs Contain Non-Numeric Types")
            
        
        if(X.shape[0] != y.shape[0]):
            raise ValueError("Mismatch in dimensions between X and y")
            

    def validate_predict_(self, X):
        
        if not hasattr(self, "beta_") :
            raise RuntimeError("Cannot run predict() before fit()") 

        if (not isinstance(X , np.ndarray)) or (not isinstance(self.beta_ , np.ndarray)) or (not np.issubdtype(X.dtype, np.number)) or (not np.issubdtype(self.beta_.dtype, np.number)):
            raise TypeError("Expected a numerical NumPy array as inputs")

    
    def insert_intercept_(self, X):
        
         return np.insert(X, 0, 1, axis=1)

        
    def solve_(self, X, y):
        
        raise NotImplementedError("solve_() method needs to be implemented")
        
```

### 5.3 Using the closed form solution for $\beta_{\mathrm{ols}}$:

#### a.Theory:
Using the OLS result:
$$
(X^TX)\hat{\beta_{\mathrm{ols}}} =X^Ty 
$$
By setting:
$$
\begin{aligned}
A &= X^TX \\
b &= X^Ty \\
\end{aligned}
$$
We get The Matrix Form of the Linear Sysmte of Equations:
$$
Ax = b
$$
By LU decomposition of $A$:
$$
Ax = b \Longrightarrow \quad LUx = b \\
$$
Which we can divide into two equations:
$$
\begin{aligned}
Ly &= b \\
Ux &= y 
\end{aligned}
$$
Wich makes it easier to solve by exploiting the traingular forms of both L and U by using the following algorithms
- **Forward Substition**: For: $ Ly = b \ $
- **Backward Substition**: For $Ux = y $
#### b.Complexity: 
$$
O(p^2n + 2p^3/3)
$$
Which is faster assympotitically then calculating the inverse of matrix which has a complexity of:
$$
O(p^3)
$$

#### c. Implementation

```python
class OrdinaryOLS(BaseOLS):
    def solve_(self, X, y):
        A = X.T @ X
        b =  X.T @ y
        return np.linalg.solve(A, b) 
```

### 5.4 Using Cholesky Decomposition

#### a.Approach:
Instead of calculating $\hat{ \beta_{\mathrm{ols}}} $ usign the closed form solution 
It can be calculated as a solution to the equation:
$$
(X^TX)\hat{ \beta_{\mathrm{ols}}} = X^Ty
$$
which is of the form:
$$
Ax = b
$$
Which has a nicer way of solving using Cholesky Decomposition since $X^TX$ is symmetric postive definite:
$$
Ax = b \Longrightarrow \quad LL^Tx = b \\
$$
where L is a Lower Triangular Matrix  
by setting: 
$$
z = L^Tx 
$$
then we can solve each equation alone by forward and backward substitions exploiting the triangular form proeprty of the L matrix.

#### b.Complexity: 
We have:
- Calculating the Product of: $\ \ X^T  \ \in \mathbb{R}^{(p+1) \times n} \ $ and  $ \ X \in \mathbb{R}^{n \times (p+1)} \ $: $O(p^2n)$
- Calculating the Product of: $ \ X^T \in \mathbb{R}^{(p+1) \times n}$ and $y \in \mathbb{R}^{n} \ $ : $O(n^p)$
- Cholesky Factorization:  $ \ X^TX \in \mathbb{R}^{(p+1) \times (p+1)} \ \ $ : $ O(\frac{p^3}{3}) $
- Forward substition for:  $ Lz = X^Ty  $ : $O(p^2)$
- backward substition for: $ L^T\beta = z$ : $O(p^2)$

$$ \\[1em] $$
By summing:
$$
O(p^2n + 3 p^3/3)
$$


#### c.Properties
**Advantages:**
- Numerical stability
- Small Constant Factors
- More Efficient Memory Usage
    
**Limitations:**  
- Not stable when X is near Singular
  

#### d. Implementation

```python
class CholeskyOLS(BaseOLS):
    def solve_(self, X, y):
        A = X.T @ X
        b = X.T @ y
        L = np.linalg.cholesky(A)
        z = np.linalg.solve(L, b)
        return np.linalg.solve(L.T, z)
```

### 5.5 Using QR Decomposition 

#### a. Approach
By using QR decompostion:
$$
X = QR
$$
Where $Q \in \mathbb{R}^{n \times (p+1)}$ is an Orthonormal(Rotation) Matrix  
And $ R \in \mathbb{R}^{(p+1) \times (p+1)}$ is an Upper Triangular Matrix
Which can be calculated using Gram-Schmidt Algorithm which is just an iterative orthogonal projection process but modern alogrithms HouseHolder QR are considered better


$$ \\ $$
by using :
$$
Q^TQ = I_p \ \ \text{and} \ \ QQ^T = I_n
$$
We get:
$$
\begin{aligned}
(X^TX)\beta = X^Ty &\Longrightarrow R^TR\beta = R^TQ^Ty \\
&\Longrightarrow R\beta = Q^Ty
\end{aligned}
$$
Which can be easily solved using Backward substition

#### b.Complexity: 
By using householder QR 
$$
\begin{aligned}
\text{QR decomposition} &: \mathcal{O}(np^2),\\
Q^\top y &: \mathcal{O}(np),\\
\text{Backward substitution} &: \mathcal{O}(p^2).
\end{aligned}
$$

#### c.Properties

#### d.Implementation

```python
class QrOLS(BaseOLS):
    def solve_(self, X, y):
        Q, R = np.linalg.qr(X)
        b = Q.T @ y
        return np.linalg.solve(R, b)
```

### 5.6 Gradient Descent Implementation

#### a.Approach
An iterative algorithm for finding $\beta$ by tweaking the value of $\beta$ using this formula:
$$
\beta^{(k+1)} = \beta^{(k)} - \eta \nabla_{\beta}RSS(\beta^{(k)})
$$
where:
- $ \beta^{(k)} \ $: is value of \beta at iteration k 
- $ \eta $: is the learning rate (how big the steps are relative to the value of the gradient)  
- $ \nabla_{\beta}RSS(\beta^{(k)})$ is the gradient of the Residual Sum of Squares with respect to: $ \beta $ 

from previous results we have:
$$
\nabla_{\beta}RSS(\beta^{(k)}) = -2X^T(y - X\beta^{(k)})
$$
thus 
$$
\beta^{(k+1)} = \beta^{(k)} +  2\eta X^T(y - X\beta^{(k)})
$$
And since we assumed $X^TX$ is positive definite  it means the Hessian matrix of RSS is positive definite 
thus the RSS is  a Quadratic strictly Convex Function meaning it has one **Unique Global Minimum** 
Therefore choosing the value of $\beta_0$ doesn't matter as long as a good learning rate $\eta$ is chosen

Choosing $\eta$ :
$$
\begin{aligned}
e_{k+1} &= \beta_{k+1} - \beta \\
&= g(\beta_{k}) - g(\beta) \\
&= g(\beta + e_{k}) - g(\beta) \\
&= g(\beta) + g^-(\beta)e_{k}) - g(\beta) \\
&= g^-(\beta)e_{k} \\
&= (I - 2\eta X^TX)e_{k}
\end{aligned}
$$

So we have:
$$
\begin{aligned}
\Longrightarrow e_n = (I - 2\eta X^TX)^n \ e_0
\end{aligned}
$$

which converges when 
$$
\begin{aligned}
\lim_{n \to +\infty} (I - 2\eta X^TX)^n = 0
\end{aligned}
$$
Let $\lambda_i$ be the eigen values of X^TX
We have:
$$
\begin{aligned}
X^TX &= PDP^T  \\
\Longrightarrow I - 2\eta X^TX &= P(I - 2\eta D) P^T \\
&= PD'P^T
\end{aligned}
$$
where $D'$ is a diagonal matrix with $\lambda_i'$ as enteries where:
$$
\lambda_i' = 1 - 2\eta \lambda_i
$$

then 
$$
(I - 2\eta X^TX)^n = P(D')^nP
$$
Which converges only if $(D')^n$ converges  
Which Converges iff
$$
\begin{aligned}
\lim_{n \to \infty} \lambda_i' &= 0  \ \ \text{Where} \ \ i \in \{1 \dots n\} \\
\Longrightarrow  \lim_{n \to \infty} (1 - 2\eta \lambda_i)^n &= 0  \\
\Longrightarrow  \ \  -1 < 1 - 2\eta \lambda_i &< 1 \\
\Longrightarrow \ \ 0 < \eta &< \frac{2}{\lambda_i} \\
\Longrightarrow \ \ 0 < \eta &< \frac{2}{\lambda_{max}}
\end{aligned}
$$
